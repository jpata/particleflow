```python
from postprocessing_hits import process_one_file
import uproot

import os

dataset = "fcc"
out_directory = dataset+"_data" + "/parquet"

os.makedirs(out_directory, exist_ok=True)

fi = dataset+"_data"+"/reco_p8_ee_tt_ecm365_20414.root"
fo = out_directory+"/reco_p8_ee_tt_ecm365_20414.parquet"

out = process_one_file(fi, fo, dataset)
```

```python
import sys

sys.path.append(os.path.abspath(os.path.join(os.path.dirname("/Users/michaelzhang/Desktop/projects/particleflow/mlpf/data/key4hep"), "..")))

from heptfds.cld_pf_edm4hep_hits.utils_edm import generate_examples

gen = generate_examples([fo,])
d = next(gen)[1]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pdgids = list(d["ygen"][d["ygen"][:, 0] != 0.0][:,0].astype(int))
print("number of gen particles: ", len(pdgids))

start = 20
end = 40
pdgids = pdgids[start:end]

plt.figure(figsize=(12, 5))

# plt.imshow(d["gp_to_hits"], cmap='gray_r')
sns.heatmap(d["gp_to_hits"][start:end], cmap="coolwarm", cbar=True, xticklabels=False, yticklabels=pdgids)

# plt.colorbar()
plt.title(dataset+' hit association matrix')
plt.ylabel('gen particle')
plt.xlabel('associated hits')
plt.show()

plt.figure(figsize=(12, 5))

# plt.imshow(d["gp_to_hits"], cmap='gray_r')
sns.heatmap(d["gp_to_tracks"][start:end], cmap="gray_r", cbar=True, xticklabels=True, yticklabels=pdgids, linewidths=1, linecolor="black")

# plt.colorbar()
plt.title(dataset+' track association matrix')
plt.ylabel('gen particle')
plt.xlabel('associated tracks')
plt.show()
```
